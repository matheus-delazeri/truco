## Ajustes Gerais

- Colunas 'bot' alteradas para 'opponent'
- Colunas 'player' alteradas para 'self'
- Colunas 'play' alteradas para 'round' e valores agora começam do index 0 (ao invés de `_play_1`, por exemplo, é `_round_0`)

```python
import pandas as pd
import numpy as np

df = pd.read_csv("cases_old.csv")
```

## Ajuste nas jogadas do oponente

- Ao invés de exibir o valor da carta jogada, exibe o ID do ranking da mesma. Sendo:
-- Low card -> 9
-- Medium card -> 10
-- High card -> 11

```python
df.loc[df['opponent_round_0'] == df['opponent_low_card'], 'opponent_round_0'] = 9
df.loc[df['opponent_round_1'] == df['opponent_low_card'], 'opponent_round_1'] = 9
df.loc[df['opponent_round_2'] == df['opponent_low_card'], 'opponent_round_2'] = 9


df.loc[df['opponent_round_1'] == df['opponent_medium_card'], 'opponent_round_1'] = 10
df.loc[df['opponent_round_2'] == df['opponent_medium_card'], 'opponent_round_2'] = 10
df.loc[df['opponent_round_0'] == df['opponent_medium_card'], 'opponent_round_0'] = 10

df.loc[df['opponent_round_1'] == df['opponent_high_card'], 'opponent_round_1'] = 11
df.loc[df['opponent_round_2'] == df['opponent_high_card'], 'opponent_round_2'] = 11
df.loc[df['opponent_round_0'] == df['opponent_high_card'], 'opponent_round_0'] = 11

df[['opponent_round_1', 'opponent_round_2', 'opponent_round_0']]
```

## Ajuste nas jogadas do jogador principal

- Ao invés de exibir o valor da carta jogada, exibe o ID do ranking da mesma. Sendo:
-- Low card -> 9
-- Medium card -> 10
-- High card -> 11

```python
df.loc[df['self_round_1'] == df['self_low_card'], 'self_round_1'] = 9
df.loc[df['self_round_2'] == df['self_low_card'], 'self_round_2'] = 9
df.loc[df['self_round_0'] == df['self_low_card'], 'self_round_0'] = 9


df.loc[df['self_round_1'] == df['self_medium_card'], 'self_round_1'] = 10
df.loc[df['self_round_2'] == df['self_medium_card'], 'self_round_2'] = 10
df.loc[df['self_round_0'] == df['self_medium_card'], 'self_round_0'] = 10

df.loc[df['self_round_1'] == df['self_high_card'], 'self_round_1'] = 11
df.loc[df['self_round_2'] == df['self_high_card'], 'self_round_2'] = 11
df.loc[df['self_round_0'] == df['self_high_card'], 'self_round_0'] = 11

df[['self_round_1', 'self_round_2', 'self_round_0']]
```

## Altera o valor `-46` e `-1` por `fold` 

Segundo o mapeamento fornecido, o valor `-46` representa que o adversário foi ao baralho e
`-1` que uma carta foi virada.
Neste caso, não será necessário utilizar estas informações.

```python
df.loc[df['opponent_round_1'].isin([-46, -1]), 'opponent_round_1'] = 8
df.loc[df['opponent_round_2'].isin([-46, -1]), 'opponent_round_2'] = 8
df.loc[df['opponent_round_0'].isin([-46, -1]), 'opponent_round_0'] = 8

df.loc[df['self_round_1'].isin([-46, -1]), 'self_round_1'] = 8
df.loc[df['self_round_2'].isin([-46, -1]), 'self_round_2'] = 8
df.loc[df['self_round_0'].isin([-46, -1]), 'self_round_0'] = 8

df[['self_round_1', 'self_round_2', 'self_round_0', 'opponent_round_1', 'opponent_round_2', 'opponent_round_0']]
```

```python
df.loc[df['opponent_round_1'].isin([-15]), 'opponent_round_1'] = 8
df.loc[df['opponent_round_2'].isin([-15]), 'opponent_round_2'] = 8
df.loc[df['opponent_round_0'].isin([-15]), 'opponent_round_0'] = 8

df.loc[df['self_round_1'].isin([-15]), 'self_round_1'] = 8
df.loc[df['self_round_2'].isin([-15]), 'self_round_2'] = 8
df.loc[df['self_round_0'].isin([-15]), 'self_round_0'] = 8

df[['self_round_1', 'self_round_2', 'self_round_0', 'opponent_round_1', 'opponent_round_2', 'opponent_round_0']]
```
